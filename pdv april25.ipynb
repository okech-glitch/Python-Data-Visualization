```python
import pandas as pd
import matplotlib.pyplot as plt

percentage = [30, 20, 15, 10, 5]
dept       = ['IT', 'HR', 'Finance', 'Production', 'Accounts']
color      = ['red', 'yellow', 'green', 'orange', 'blue']
explode    = [0.1, 0.2, 0.1, 0.1, 0.1]

plt.pie(percentage, labels=dept, colors=color, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Budgetary Allocation Pie Chart")
plt.legend(title="Departments", loc="upper left")
plt.axis('equal')

plt.show()
```

```python
QuarterData = {
    'Quarter': ['Q1_2023', 'Q2_2023', 'Q3_2023', 'Q4_2023', 
                'Q1_2024', 'Q2_2024', 'Q3_2024', 'Q4_2024'],
    'Sales': [2500, 2700, 2740, 3200, 3700, 6300, 8100, 1150],
    'Customers': [290, 300, 305, 680, 1850, 2850, 3500, 3800],
    'Marketing_Spend': [300, 400, 450, 750, 950, 1450, 1850, 3000]
}
df = pd.DataFrame(QuarterData)

plt.figure(figsize=(10,6))
plt.plot(df['Quarter'], df['Sales'], marker='o', label='Sales', color='blue')
plt.plot(df['Quarter'], df['Customers'], marker='s', label='Customers', color='green')
plt.plot(df['Quarter'], df['Marketing_Spend'], marker='^', label='Marketing Spend', color='red')

plt.title('Sales, Customers and Marketing Spend Over Time', fontsize=14)
plt.xlabel('Quarter')
plt.ylabel('Amount / Number')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.pie(df['Marketing_Spend'], labels=df['Quarter'], autopct='%1.1f%%', startangle=90)
plt.title("Marketing Expenditure Distribution by Quarter")
plt.axis('equal')
plt.show()
```

```python
correlation = df[['Sales', 'Customers', 'Marketing_Spend']].corr()
print(correlation)

# Optional: visualize with heatmap (bonus marks!)
import seaborn as sns
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title("Variables Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Marketing_Spend'], df['Sales'], color='purple', s=100)
plt.title("Effect of Marketing Spend on Sales")
plt.xlabel("Marketing Spend")
plt.ylabel("Sales")
plt.grid(True)

# Optional: add trend line
import numpy as np
z = np.polyfit(df['Marketing_Spend'], df['Sales'], 1)
p = np.poly1d(z)
plt.plot(df['Marketing_Spend'], p(df['Marketing_Spend']), "r--")

plt.show()
```
